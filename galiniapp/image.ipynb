```python
import matplotlib.pyplot as plt

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Define the boxes and their positions
components = {
    "Client Side (Frontend)": (1, 3),
    "Server Side (Backend)": (1, 2),
    "Database": (1, 1),
    "External Services": (3, 1)
}

# Draw the boxes
for component, (x, y) in components.items():
    ax.text(x, y, component, fontsize=12, bbox=dict(facecolor='cyan', edgecolor='black', boxstyle='round,pad=0.5'), ha='center', va='center')

# Draw arrows
ax.annotate("", xy=(1, 2.5), xytext=(1, 2.8), arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(1, 1.5), xytext=(1, 1.8), arrowprops=dict(arrowstyle="->"))
ax.annotate("", xy=(2, 1), xytext=(1.2, 1), arrowprops=dict(arrowstyle="->"))

# Draw the descriptions inside the boxes
descriptions = {
    "Client Side (Frontend)": "- User Interface (React or Django Templates)\n- Requests to Backend",
    "Server Side (Backend)": "- Django Framework\n- Business Logic\n- Request Handling",
    "Database": "- SQLite (db.sqlite3)\n- User Data, Appointments,\n  Quizzes, Video Chats, SMS,\n  Hospital Locations",
    "External Services": "- Google Translate\n- ZegoCloud (Video Chat)\n- SMS Gateway\n- Map Service"
}

# Add descriptions
for component, description in descriptions.items():
    x, y = components[component]
    ax.text(x, y - 0.3, description, fontsize=10, ha='center', va='center')

# Set limits and hide the axes
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.axis('off')

# Save the figure
plt.savefig('/mnt/data/galini_plus_system_architecture.png')
plt.show()
```
